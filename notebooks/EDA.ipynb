```python
X = df.drop('Concrete compressive strength(MPa, megapascals) ' ,axis=1)
y = df['Concrete compressive strength(MPa, megapascals) ']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train,  y_test = train_test_split(X , y , test_size=0.25 , random_state=0)
```

```python
from sklearn.linear_model import LinearRegression , Lasso , Ridge , ElasticNet
```

```python
from sklearn.ensemble import RandomForestRegressor , GradientBoostingRegressor
from sklearn.metrics import r2_score , mean_squared_error
```

```python
model1 = LinearRegression()
model2 = Lasso()
model3 = Ridge()
model4 = ElasticNet()
model6 = RandomForestRegressor()
model7 = GradientBoostingRegressor()
```

```python
model1.fit(X_train , y_train)
model2.fit(X_train , y_train)
model3.fit(X_train , y_train)
model4.fit(X_train , y_train)
model6.fit(X_train , y_train)
model7.fit(X_train , y_train)
```

```python
score_1 = model1.score(X_train , y_train)
score_2 = model2.score(X_train , y_train)
score_3 = model3.score(X_train , y_train)
score_4 = model4.score(X_train , y_train)
score_6 = model6.score(X_train , y_train)
score_7 = model7.score(X_train , y_train)
```

```python
print(f'LinearRegression : {score_1} ,Lasso : {score_2} ,Ridge : {score_3} ,ElasticNet : {score_4} ,RandomForestRegressor : {score_6} ,GradientBoostingRegressor : {score_7}')
```

```python
y_pred_1 = model1.predict(X_test)
y_pred_2 = model2.predict(X_test)
y_pred_3 = model3.predict(X_test)
y_pred_4 = model4.predict(X_test)
y_pred_6 = model6.predict(X_test)
y_pred_7 = model7.predict(X_test)
```

```python
mse1 = mean_squared_error(y_test , y_pred_1)
mse2 = mean_squared_error(y_test , y_pred_2)
mse3 = mean_squared_error(y_test , y_pred_3)
mse4 = mean_squared_error(y_test , y_pred_4)
mse6 = mean_squared_error(y_test , y_pred_6)
mse7 = mean_squared_error(y_test , y_pred_7)
```

```python
print(f'LinearRegression : {mse1} ,Lasso : {mse2} ,Ridge : {mse3} ,ElasticNet : {mse4} ,RandomForestRegressor : {mse6} ,GradientBoostingRegressor : {mse7}')
```

```python
r2_score_1 = r2_score(y_test , y_pred_1)
r2_score_2 = r2_score(y_test , y_pred_2)
r2_score_3 = r2_score(y_test , y_pred_3)
r2_score_4 = r2_score(y_test , y_pred_4)
r2_score_6 = r2_score(y_test , y_pred_6)
r2_score_7 = r2_score(y_test , y_pred_7)
```

```python
print(f'LinearRegression : {r2_score_1} ,Lasso : {r2_score_2} ,Ridge : {r2_score_3} ,ElasticNet : {r2_score_4} ,RandomForestRegressor : {r2_score_6} ,GradientBoostingRegressor : {r2_score_7}')
```
